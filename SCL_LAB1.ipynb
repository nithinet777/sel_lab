```python
#5genetic algorithm
import random

def fitness(x):
    return -(x - 5) ** 2 + 25

def create_individual():
    return random.uniform(-10, 10)

def mutate(x, rate=0.1):
    if random.random() < rate:
        x += random.uniform(-1, 1)
    return x

def crossover(a, b):
    t = random.random()
    return t * a + (1 - t) * b

def select(pop):
    k = 3
    c = random.sample(pop, k)
    return max(c, key=fitness)

def genetic_algorithm(pop_size=50, generations=10):
    pop = [create_individual() for _ in range(pop_size)]

    for gen in range(generations):
        
        best = max(pop, key=fitness)
        print(f"Generation {gen}: Best = {best:.4f}, Fitness = {fitness(best):.4f}")

        new_pop = []
        for _ in range(pop_size):
            p1 = select(pop)
            p2 = select(pop)
            child = crossover(p1, p2)
            child = mutate(child)
            new_pop.append(child)

        pop = new_pop

    best = max(pop, key=fitness)
    return best

best = genetic_algorithm()
print("\nFinal Best:", best, "Fitness:", fitness(best))
```

```python
#4)particle swarm intelligence
import random

def fitness(x):
    return -(x - 5) ** 2 + 25


def create_particle():
    position = random.uniform(-10, 10)
    velocity = random.uniform(-1, 1)
    return position, velocity


def particle_swarm_optimization(pop_size=50, generations=10):

    particles = []
    velocities = []
    pbest = []

    for _ in range(pop_size):
        pos, vel = create_particle()
        particles.append(pos)
        velocities.append(vel)
        pbest.append(pos)

    gbest = max(particles, key=fitness)

    w = 0.5
    c1 = 1.5
    c2 = 1.5

    for gen in range(generations):

        for i in range(pop_size):

            r1 = random.random()
            r2 = random.random()

            velocities[i] = (
                w * velocities[i]
                + c1 * r1 * (pbest[i] - particles[i])
                + c2 * r2 * (gbest - particles[i])
            )

            particles[i] = particles[i] + velocities[i]

            particles[i] = max(-10, min(10, particles[i]))

            if fitness(particles[i]) > fitness(pbest[i]):
                pbest[i] = particles[i]

        gbest = max(pbest, key=fitness)

        print(f"Generation {gen}: Best = {gbest:.4f}, Fitness = {fitness(gbest):.4f}")

    return gbest


best = particle_swarm_optimization()

print("\nFinal Best:", best, "Fitness:", fitness(best))
```

```python
#grey wolf 
import random

def fitness(x):
    return -(x - 5) ** 2 + 25


def create_wolf():
    return random.uniform(-10, 10)


def update_position(wolf, alpha, beta, delta, a):

    r1 = random.random()
    r2 = random.random()
    A1 = 2 * a * r1 - a
    C1 = 2 * r2
    D_alpha = abs(C1 * alpha - wolf)
    X1 = alpha - A1 * D_alpha

    r1 = random.random()
    r2 = random.random()
    A2 = 2 * a * r1 - a
    C2 = 2 * r2
    D_beta = abs(C2 * beta - wolf)
    X2 = beta - A2 * D_beta

    r1 = random.random()
    r2 = random.random()
    A3 = 2 * a * r1 - a
    C3 = 2 * r2
    D_delta = abs(C3 * delta - wolf)
    X3 = delta - A3 * D_delta

    new_pos = (X1 + X2 + X3) / 3
    return new_pos


def grey_wolf_optimizer(pop_size=50, generations=10):

    wolves = [create_wolf() for _ in range(pop_size)]

    for gen in range(generations):

        wolves.sort(key=fitness, reverse=True)

        alpha = wolves[0]
        beta = wolves[1]
        delta = wolves[2]

        print(f"Generation {gen}: Alpha = {alpha:.4f}, Fitness = {fitness(alpha):.4f}")

        a = 2 - gen * (2 / generations)

        new_wolves = []

        for wolf in wolves:
            new_pos = update_position(wolf, alpha, beta, delta, a)

            new_pos = max(-10, min(10, new_pos))

            new_wolves.append(new_pos)

        wolves = new_wolves

    best = max(wolves, key=fitness)
    return best


best = grey_wolf_optimizer()

print("\nFinal Best:", best, "Fitness:", fitness(best))


```

```python
#3)ANT COLONY
import random

def fitness(x):
    return -(x - 5) ** 2 + 25


def create_ant():
    return random.uniform(-10, 10)


def ant_colony(pop_size=50, generations=10):

    ants = [create_ant() for _ in range(pop_size)]
    pheromone = [1.0] * pop_size

    evaporation = 0.5
    Q = 1

    for gen in range(generations):

        for i in range(pop_size):
            step = random.uniform(-1, 1)
            ants[i] = ants[i] + step

            ants[i] = max(-10, min(10, ants[i]))

        for i in range(pop_size):
            pheromone[i] = (1 - evaporation) * pheromone[i]

            pheromone[i] += Q * fitness(ants[i])

        best = max(ants, key=fitness)

        print(f"Generation {gen}: Best = {best:.4f}, Fitness = {fitness(best):.4f}")

    best = max(ants, key=fitness)
    return best


best = ant_colony()

print("\nFinal Best:", best, "Fitness:", fitness(best))
```

```python
#artificial bee keep

import random

def fitness(x):
    return -(x - 5) ** 2 + 25


def create_food():
    return random.uniform(-10, 10)


def neighbor(x):
    phi = random.uniform(-1, 1)
    k = create_food()
    return x + phi * (x - k)


def artificial_bee_colony(pop_size=50, generations=10):

    food_sources = [create_food() for _ in range(pop_size)]
    trials = [0] * pop_size
    limit = 10

    for gen in range(generations):

        # ----- Employed Bees Phase -----
        for i in range(pop_size):
            new = neighbor(food_sources[i])

            if fitness(new) > fitness(food_sources[i]):
                food_sources[i] = new
                trials[i] = 0
            else:
                trials[i] += 1

        # ----- Onlooker Bees Phase -----
        total_fit = sum(fitness(x) for x in food_sources)

        for i in range(pop_size):
            prob = fitness(food_sources[i]) / total_fit

            if random.random() < prob:
                new = neighbor(food_sources[i])

                if fitness(new) > fitness(food_sources[i]):
                    food_sources[i] = new
                    trials[i] = 0
                else:
                    trials[i] += 1

        # ----- Scout Bees Phase -----
        for i in range(pop_size):
            if trials[i] > limit:
                food_sources[i] = create_food()
                trials[i] = 0

        best = max(food_sources, key=fitness)
        print(f"Generation {gen}: Best = {best:.4f}, Fitness = {fitness(best):.4f}")

    best = max(food_sources, key=fitness)
    return best


best = artificial_bee_colony()

print("\nFinal Best:", best, "Fitness:", fitness(best))
```

```python
import random
import math

def fitness(x):
    return -(x - 5) ** 2 + 25


def create_firefly():
    return random.uniform(-10, 10)


def distance(a, b):
    return abs(a - b)


def move_firefly(xi, xj, beta0=1, gamma=1, alpha=0.2):
    
    r = distance(xi, xj)
    
    beta = beta0 * math.exp(-gamma * r * r)
    
    new_pos = xi + beta * (xj - xi) + alpha * (random.random() - 0.5)
    
    return new_pos


def firefly_algorithm(pop_size=50, generations=10):

    fireflies = [create_firefly() for _ in range(pop_size)]

    for gen in range(generations):

        for i in range(pop_size):
            for j in range(pop_size):

                if fitness(fireflies[j]) > fitness(fireflies[i]):

                    new_pos = move_firefly(fireflies[i], fireflies[j])

                    # keep within bounds
                    new_pos = max(-10, min(10, new_pos))

                    fireflies[i] = new_pos

        best = max(fireflies, key=fitness)
        print(f"Generation {gen}: Best = {best:.4f}, Fitness = {fitness(best):.4f}")

    best = max(fireflies, key=fitness)
    return best


best = firefly_algorithm()

print("\nFinal Best:", best, "Fitness:", fitness(best))
```
